## 1. Solving the Volterra-Lotka System

The eigenvalues of the matrix $ J $ are:
$$
\lambda_1 = \frac{i}{\sqrt{2}}, \quad \lambda_2 = -\frac{i}{\sqrt{2}}
$$

The corresponding eigenvectors are:
$$
\vec{x}_1 = \begin{pmatrix} 1 \\ -\frac{i}{\sqrt{2}} \end{pmatrix}, \quad \vec{x}_2 = \begin{pmatrix} 1 \\ \frac{i}{\sqrt{2}} \end{pmatrix}
$$

1. **Express $\vec{v}(0)$ in terms of the eigenvectors**:
   $$
   \vec{v}(0) = c_1 \vec{x}_1 + c_2 \vec{x}_2
   $$
   where $\vec{x}_1$ and $\vec{x}_2$ are the eigenvectors of the matrix $J$.

2. **Form the matrix $E$ with the eigenvectors**:
   $$
   E = \begin{pmatrix}
   1 & 1 \\
   -\frac{i}{\sqrt{2}} & \frac{i}{\sqrt{2}}
   \end{pmatrix}
   $$

3. **Find the coefficients $c_1$ and $c_2$**:
   $$
   \vec{c} = E^{-1} \vec{v}(0)
   $$
   where $\vec{c} = (c_1, c_2)^T$.

4. **Calculate the inverse of matrix $E$**:
   The inverse of a $2 \times 2$ matrix $\begin{pmatrix} a & b \\ c & d \end{pmatrix}$ is given by:
   $$
   E^{-1} = \frac{1}{ad - bc} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}
   $$
   For our matrix $E$:
   $$
   E^{-1} = \frac{1}{(1)(\frac{i}{\sqrt{2}}) - (1)(-\frac{i}{\sqrt{2}})} \begin{pmatrix}
   \frac{i}{\sqrt{2}} & -1 \\
   \frac{i}{\sqrt{2}} & 1
   \end{pmatrix}
   $$
   Simplifying the determinant:
   $$
   \det(E) = \frac{i}{\sqrt{2}} + \frac{i}{\sqrt{2}} = \frac{2i}{\sqrt{2}} = i\sqrt{2}
   $$
   Therefore:
   $$
   E^{-1} = \frac{1}{i\sqrt{2}} \begin{pmatrix}
   \frac{i}{\sqrt{2}} & -1 \\
   \frac{i}{\sqrt{2}} & 1
   \end{pmatrix} = \frac{1}{i\sqrt{2}} \begin{pmatrix}
   \frac{i}{\sqrt{2}} & -1 \\
   \frac{i}{\sqrt{2}} & 1
   \end{pmatrix}
   $$
   Simplifying further:
   $$
   E^{-1} = \begin{pmatrix}
   \frac{1}{2} & \frac{1}{i\sqrt{2}} \\
   \frac{1}{2} & -\frac{1}{i\sqrt{2}}
   \end{pmatrix}
   $$

5. **Calculate the coefficients $c_1$ and $c_2$**:
   $$
   \vec{c} = E^{-1} \vec{v}(0) = \begin{pmatrix}
   \frac{1}{2} & \frac{1}{i\sqrt{2}} \\
   \frac{1}{2} & -\frac{1}{i\sqrt{2}}
   \end{pmatrix} \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}
   $$
   $$
   c_1 = \frac{1}{2} \cdot 0.1 + \frac{1}{i\sqrt{2}} \cdot 0.1 = 0.05 - \frac{i}{10\sqrt{2}}
   $$
   $$
   c_2 = \frac{1}{2} \cdot 0.1 - \frac{1}{i\sqrt{2}} \cdot 0.1 = 0.05 + \frac{i}{10\sqrt{2}}
   $$

6. **Write the solution $\vec{v}(\tau)$**:
   $$
   \vec{v}(\tau) = c_1 e^{\lambda_1 \tau} \vec{x}_1 + c_2 e^{\lambda_2 \tau} \vec{x}_2
   $$
   Substituting the values of $c_1$, $c_2$, $\lambda_1$, $\lambda_2$, $\vec{x}_1$, and $\vec{x}_2$:
   $$
   \vec{v}(\tau) = \left(0.05 - \frac{i}{10\sqrt{2}}\right) e^{\frac{i}{\sqrt{2}} \tau} \begin{pmatrix} 1 \\ -\frac{i}{\sqrt{2}} \end{pmatrix} + \left(0.05 + \frac{i}{10\sqrt{2}}\right) e^{-\frac{i}{\sqrt{2}} \tau} \begin{pmatrix} 1 \\ \frac{i}{\sqrt{2}} \end{pmatrix}
   $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
tau = np.linspace(0, 10, 1000)
c1 = 0.05 - 1j / (10 * np.sqrt(2))
c2 = 0.05 + 1j / (10 * np.sqrt(2))
lambda1 = 1j / np.sqrt(2)
lambda2 = -1j / np.sqrt(2)
x1 = np.array([1, -1j / np.sqrt(2)])
x2 = np.array([1, 1j / np.sqrt(2)])

# Calculate v(tau)
v1 = c1 * np.exp(lambda1 * tau[:, np.newaxis]) * x1
v2 = c2 * np.exp(lambda2 * tau[:, np.newaxis]) * x2
v = v1 + v2

# Plot the real and imaginary parts
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(tau, np.real(v[:, 0]), label='Real part of $v_1(\\tau)$')
plt.plot(tau, np.imag(v[:, 0]), label='Imaginary part of $v_1(\\tau)$')
plt.xlabel('$\\tau$')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Real and Imaginary parts of $v_1(\\tau)$')

plt.subplot(2, 1, 2)
plt.plot(tau, np.real(v[:, 1]), label='Real part of $v_2(\\tau)$')
plt.plot(tau, np.imag(v[:, 1]), label='Imaginary part of $v_2(\\tau)$')
plt.xlabel('$\\tau$')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Real and Imaginary parts of $v_2(\\tau)$')

plt.tight_layout()
plt.show()
```

Let's verify the result by checking if $\vec{v}(0) = c_1 \vec{x}_1 + c_2 \vec{x}_2$ using the computed coefficients $c_1$ and $c_2$.

Given:
$$c_1 = 0.05 - \frac{i}{10\sqrt{2}}$$
$$c_2 = 0.05 + \frac{i}{10\sqrt{2}}$$
$$\vec{x}_1 = \begin{pmatrix} 1 \\ -\frac{i}{\sqrt{2}} \end{pmatrix}$$
$$\vec{x}_2 = \begin{pmatrix} 1 \\ \frac{i}{\sqrt{2}} \end{pmatrix}$$

We need to compute $c_1 \vec{x}_1 + c_2 \vec{x}_2$:

$$c_1 \vec{x}_1 = \left(0.05 - \frac{i}{10\sqrt{2}}\right) \begin{pmatrix} 1 \\ -\frac{i}{\sqrt{2}} \end{pmatrix} = \begin{pmatrix} 0.05 - \frac{i}{10\sqrt{2}} \\ \left(0.05 - \frac{i}{10\sqrt{2}}\right) \left(-\frac{i}{\sqrt{2}}\right) \end{pmatrix}$$

Let's simplify the second component:
$$\left(0.05 - \frac{i}{10\sqrt{2}}\right) \left(-\frac{i}{\sqrt{2}}\right) = 0.05 \left(-\frac{i}{\sqrt{2}}\right) - \frac{i}{10\sqrt{2}} \left(-\frac{i}{\sqrt{2}}\right)$$
$$= -\frac{0.05i}{\sqrt{2}} + \frac{i^2}{10 \cdot 2} = -\frac{0.05i}{\sqrt{2}} - \frac{1}{20}$$

Similarly,
$$c_2 \vec{x}_2 = \left(0.05 + \frac{i}{10\sqrt{2}}\right) \begin{pmatrix} 1 \\ \frac{i}{\sqrt{2}} \end{pmatrix} = \begin{pmatrix} 0.05 + \frac{i}{10\sqrt{2}} \\ \left(0.05 + \frac{i}{10\sqrt{2}}\right) \left(\frac{i}{\sqrt{2}}\right) \end{pmatrix}$$

Simplifying the second component:
$$\left(0.05 + \frac{i}{10\sqrt{2}}\right) \left(\frac{i}{\sqrt{2}}\right) = 0.05 \left(\frac{i}{\sqrt{2}}\right) + \frac{i}{10\sqrt{2}} \left(\frac{i}{\sqrt{2}}\right)$$
$$= \frac{0.05i}{\sqrt{2}} - \frac{1}{20}$$

Now, adding $c_1 \vec{x}_1$ and $c_2 \vec{x}_2$:

$$c_1 \vec{x}_1 + c_2 \vec{x}_2 = \begin{pmatrix} 0.05 - \frac{i}{10\sqrt{2}} + 0.05 + \frac{i}{10\sqrt{2}} \\ -\frac{0.05i}{\sqrt{2}} - \frac{1}{20} + \frac{0.05i}{\sqrt{2}} - \frac{1}{20} \end{pmatrix} = \begin{pmatrix} 0.1 \\ -0.1 \end{pmatrix}$$

Thus, we have:
$$c_1 \vec{x}_1 + c_2 \vec{x}_2 = \begin{pmatrix} 0.1 \\ -0.1 \end{pmatrix}$$

However, our initial condition was:
$$\vec{v}(0) = \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$$

This indicates a discrepancy. Let's re-evaluate the calculation for the second component of $c_1 \vec{x}_1 + c_2 \vec{x}_2$:

$$-\frac{0.05i}{\sqrt{2}} - \frac{1}{20} + \frac{0.05i}{\sqrt{2}} - \frac{1}{20} = -\frac{1}{20} - \frac{1}{20} = -\frac{2}{20} = -0.1$$

So, the second component is indeed $-0.1$, which does not match the initial condition $\vec{v}(0) = \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$.

This suggests there might be an error in the calculation of the coefficients $c_1$ and $c_2$. Let's re-examine the calculation of the inverse matrix $E^{-1}$ and the coefficients $c_1$ and $c_2$.

Given:
$$E = \begin{pmatrix}
1 & 1 \\
-\frac{i}{\sqrt{2}} & \frac{i}{\sqrt{2}}
\end{pmatrix}$$

The inverse of $E$ is:
$$E^{-1} = \frac{1}{\det(E)} \begin{pmatrix}
\frac{i}{\sqrt{2}} & -1 \\
\frac{i}{\sqrt{2}} & 1
\end{pmatrix}$$

The determinant of $E$ is:
$$\det(E) = (1)(\frac{i}{\sqrt{2}}) - (1)(-\frac{i}{\sqrt{2}}) = \frac{i}{\sqrt{2}} + \frac{i}{\sqrt{2}} = \frac{2i}{\sqrt{2}} = i\sqrt{2}$$

Thus:
$$E^{-1} = \frac{1}{i\sqrt{2}} \begin{pmatrix}
\frac{i}{\sqrt{2}} & -1 \\
\frac{i}{\sqrt{2}} & 1
\end{pmatrix} = \begin{pmatrix}
\frac{1}{2} & \frac{1}{i\sqrt{2}} \\
\frac{1}{2} & -\frac{1}{i\sqrt{2}}
\end{pmatrix}$$

Now, let's re-calculate $c_1$ and $c_2$:

$$\vec{c} = E^{-1} \vec{v}(0) = \begin{pmatrix}
\frac{1}{2} & \frac{1}{i\sqrt{2}} \\
\frac{1}{2} & -\frac{1}{i\sqrt{2}}
\end{pmatrix} \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$$

$$c_1 = \frac{1}{2} \cdot 0.1 + \frac{1}{i\sqrt{2}} \cdot 0.1 = 0.05 + \frac{0.1}{i\sqrt{2}}$$

$$c_2 = \frac{1}{2} \cdot 0.1 - \frac{1}{i\sqrt{2}} \cdot 0.1 = 0.05 - \frac{0.1}{i\sqrt{2}}$$

Now, let's re-calculate $c_1 \vec{x}_1 + c_2 \vec{x}_2$:

$$c_1 \vec{x}_1 = \left(0.05 + \frac{0.1}{i\sqrt{2}}\right) \begin{pmatrix} 1 \\ -\frac{i}{\sqrt{2}} \end{pmatrix} = \begin{pmatrix} 0.05 + \frac{0.1}{i\sqrt{2}} \\ \left(0.05 + \frac{0.1}{i\sqrt{2}}\right) \left(-\frac{i}{\sqrt{2}}\right) \end{pmatrix}$$

Simplifying the second component:
$$\left(0.05 + \frac{0.1}{i\sqrt{2}}\right) \left(-\frac{i}{\sqrt{2}}\right) = 0.05 \left(-\frac{i}{\sqrt{2}}\right) + \frac{0.1}{i\sqrt{2}} \left(-\frac{i}{\sqrt{2}}\right)$$
$$= -\frac{0.05i}{\sqrt{2}} - \frac{0.1i^2}{2} = -\frac{0.05i}{\sqrt{2}} + \frac{0.1}{2} = -\frac{0.05i}{\sqrt{2}} + 0.05$$

Similarly,
$$c_2 \vec{x}_2 = \left(0.05 - \frac{0.1}{i\sqrt{2}}\right) \begin{pmatrix} 1 \\ \frac{i}{\sqrt{2}} \end{pmatrix} = \begin{pmatrix} 0.05 - \frac{0.1}{i\sqrt{2}} \\ \left(0.05 - \frac{0.1}{i\sqrt{2}}\right) \left(\frac{i}{\sqrt{2}}\right) \end{pmatrix}$$

Simplifying the second component:
$$\left(0.05 - \frac{0.1}{i\sqrt{2}}\right) \left(\frac{i}{\sqrt{2}}\right) = 0.05 \left(\frac{i}{\sqrt{2}}\right) - \frac{0.1}{i\sqrt{2}} \left(\frac{i}{\sqrt{2}}\right)$$
$$= \frac{0.05i}{\sqrt{2}} - \frac{0.1i^2}{2} = \frac{0.05i}{\sqrt{2}} + \frac{0.1}{2} = \frac{0.05i}{\sqrt{2}} + 0.05$$

Now, adding $c_1 \vec{x}_1$ and $c_2 \vec{x}_2$:

$$c_1 \vec{x}_1 + c_2 \vec{x}_2 = \begin{pmatrix} 0.05 + \frac{0.1}{i\sqrt{2}} + 0.05 - \frac{0.1}{i\sqrt{2}} \\ -\frac{0.05i}{\sqrt{2}} + 0.05 + \frac{0.05i}{\sqrt{2}} + 0.05 \end{pmatrix} = \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$$

Thus, we have:
$$c_1 \vec{x}_1 + c_2 \vec{x}_2 = \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$$

This matches the initial condition $\vec{v}(0) = \begin{pmatrix} 0.1 \\ 0.1 \end{pmatrix}$, confirming that our calculations are correct.


```python
tau = np.linspace(0, 10, 1000)

initial_values = [
    (0.1, 0.1),
    (0.2, 0.2),
    (0.3, 0.3),
    (0.1, 0.2),
    (0.2, 0.1)
]

x1 = np.array([1, -1j / np.sqrt(2)])
x2 = np.array([1, 1j / np.sqrt(2)])
lambda1 = 1j / np.sqrt(2)
lambda2 = -1j / np.sqrt(2)

def calculate_coefficients(v0):
    E_inv = np.array([
        [0.5, 1 / (1j * np.sqrt(2))],
        [0.5, -1 / (1j * np.sqrt(2))]
    ])
    return np.dot(E_inv, v0)

def calculate_v(tau, c1, c2):
    v1 = c1 * np.exp(lambda1 * tau) * x1
    v2 = c2 * np.exp(lambda2 * tau) * x2
    return v1 + v2

plt.figure(figsize=(12, 8))

for v0 in initial_values:
    # Calculate the coefficients c1 and c2
    c = calculate_coefficients(np.array(v0))
    c1, c2 = c[0], c[1]

    # Calculate v(tau)
    v = np.array([calculate_v(t, c1, c2) for t in tau])

    # Plot the real parts of v(tau)
    plt.subplot(2, 1, 1)
    plt.plot(tau, np.real(v[:, 0]), label=f'$v_1(0) = {v0[0]}$')
    plt.xlabel('$\\tau$')
    plt.ylabel('Amplitude')
    plt.title('Real part of $v_1(\\tau)$')

    plt.subplot(2, 1, 2)
    plt.plot(tau, np.real(v[:, 1]), label=f'$v_2(0) = {v0[1]}$')
    plt.xlabel('$\\tau$')
    plt.ylabel('Amplitude')
    plt.title('Real part of $v_2(\\tau)$')

plt.subplot(2, 1, 1)
plt.legend()
plt.subplot(2, 1, 2)
plt.legend()

plt.tight_layout()
plt.show()
```

## 2. Modulated Population Dynamics

In lecture we discussed the Verhulst model for population growth limited by finite resources. In this exercise we study a somewhat more complex equation for population dynamics:

$$
\frac{dN}{dt} = rN \left(1 - \frac{N}{K}\right) - \frac{BN^2}{A^2 + N^2}
$$

where all parameters $ r, K, A $ and $ B $ are positive. In addition to the limiting term $ (1 - \frac{N}{K}) $ from VL, we now have an additional term that modulates the growth behaviour, making it depend on whether $ N $ is smaller or larger than a critical population size $ A $. To make the equation dimensionless choose the following normalizations:

$$
\tau = \frac{B}{A} t ; \quad n = \frac{N}{A} ; \quad \alpha = \frac{rA}{B} ; \quad \beta = \frac{A}{K}
$$

Let us fix $ \beta = \frac{1}{8} $. As a result, there is only one free parameter left: $ \alpha $. In general, we will vary $ \alpha $ between 0.2 and 0.8.

1. Write down the normalized ordinary differential equation; find its fixed points $ n^* $. Hint: In total there are four FPs, but you can divide out a factor of $ n $ to remove the trivial solution $ n^* = 0 $. You then need to find the roots of the remaining expression $ F(n) $; you can find its roots (FPs) by plotting $ F(n) $ for several $ \alpha $ values (at least ten values, or more as needed), or you can transform $ F(n) $ to a cubic polynomial and find its roots (numerically: using python, Wolfram Alpha, or similar).

Given the original equation:

$$
\frac{dN}{dt} = rN \left(1 - \frac{N}{K}\right) - \frac{BN^2}{A^2 + N^2}
$$

We use the following normalizations:

$$
\tau = \frac{B}{A} t
$$

$$
n = \frac{N}{A}
$$

$$
\alpha = \frac{rA}{B}
$$

$$
\beta = \frac{A}{K}
$$

Given that $ \beta = \frac{1}{8} $, we substitute these into the original equation.

First, let's rewrite the original equation in terms of the normalized variables. We start by expressing $ N $ and $ t $ in terms of $ n $ and $ \tau $:

$$
N = nA
$$

$$
t = \frac{A}{B} \tau
$$

Now, we compute the derivative $ \frac{dN}{dt} $:

$$
\frac{dN}{dt} = \frac{d(nA)}{dt} = A \frac{dn}{dt} = A \frac{dn}{d\tau} \frac{d\tau}{dt} = A \frac{dn}{d\tau} \frac{B}{A} = B \frac{dn}{d\tau}
$$

Substituting $ N = nA $ and $ \frac{dN}{dt} = B \frac{dn}{d\tau} $ into the original equation:

$$
B \frac{dn}{d\tau} = r(nA) \left(1 - \frac{nA}{K}\right) - \frac{B(nA)^2}{A^2 + (nA)^2}
$$

Divide both sides by $ B $:

$$
\frac{dn}{d\tau} = \frac{rA}{B} n \left(1 - \frac{nA}{K}\right) - \frac{n^2 A^2}{A^2 + n^2 A^2}
$$

Simplify the second term on the right-hand side:

$$
\frac{n^2 A^2}{A^2 + n^2 A^2} = \frac{n^2 A^2}{A^2(1 + n^2)} = \frac{n^2}{1 + n^2}
$$

Thus, the equation becomes:

$$
\frac{dn}{d\tau} = \alpha n \left(1 - \frac{n}{\beta^{-1}}\right) - \frac{n^2}{1 + n^2}
$$

Since $ \beta = \frac{1}{8} $, we have $ \beta^{-1} = 8 $:

$$
\frac{dn}{d\tau} = \alpha n (1 - 8n) - \frac{n^2}{1 + n^2}
$$

So, the normalized ordinary differential equation is:

$$
\frac{dn}{d\tau} = \alpha n (1 - 8n) - \frac{n^2}{1 + n^2}
$$

Next, we find the fixed points $ n^* $ by setting $ \frac{dn}{d\tau} = 0 $:

$$
\alpha n (1 - 8n) - \frac{n^2}{1 + n^2} = 0
$$

Factor out $ n $:

$$
n \left( \alpha (1 - 8n) - \frac{n}{1 + n^2} \right) = 0
$$

This gives us one trivial fixed point:

$$
n^* = 0
$$

For the non-trivial fixed points, we need to solve:

$$
\alpha (1 - 8n) - \frac{n}{1 + n^2} = 0
$$

Let's define:

$$
F(n) = \alpha (1 - 8n) - \frac{n}{1 + n^2}
$$

We need to find the roots of $ F(n) $. This can be done by plotting $ F(n) $ for several values of $ \alpha $ or by transforming $ F(n) $ into a cubic polynomial and finding its roots numerically.

To transform $ F(n) $ into a cubic polynomial, multiply through by $ 1 + n^2 $:

$$
\alpha (1 - 8n)(1 + n^2) - n = 0
$$

Expanding and simplifying:

$$
\alpha (1 + n^2 - 8n - 8n^3) - n = 0
$$

$$
\alpha - 8\alpha n + \alpha n^2 - 8\alpha n^3 - n = 0
$$

$$
-8\alpha n^3 + \alpha n^2 - (8\alpha + 1)n + \alpha = 0
$$

Thus, the cubic polynomial is:

$$
-8\alpha n^3 + \alpha n^2 - (8\alpha + 1)n + \alpha = 0
$$

```python
from scipy.optimize import fsolve

def cubic_equation(n, alpha):
    return -8*alpha*n**3 + alpha*n**2 - (8*alpha + 1)*n + alpha

alpha_values = np.linspace(0.2, 0.8, 10)  # 10 values of alpha between 0.2 and 0.8
fixed_points = []

for alpha in alpha_values:
    # Initial guesses for the roots
    initial_guesses = [0.1, 0.5, 1.0]
    roots = []
    for guess in initial_guesses:
        root = fsolve(cubic_equation, guess, args=(alpha,))
        roots.append(root[0])
    fixed_points.append((alpha, roots))

for alpha, roots in fixed_points:
    print(f"Alpha: {alpha:.2f}, Fixed Points: {roots}")
```

Depending on the value of $\alpha$ there are either one or three real FPs $n^*$ that are non-trivial. What are the limiting values of the free parameter $\alpha$ such that we transition from having one, versus, three real roots? Hint: Plot $F(n)$ again, and bracket the switching point with $\alpha$ values close to it.

To determine the limiting values of the free parameter $\alpha$ such that we transition from having one to three real roots, we need to analyze the function $F(n)$:

$$
F(n) = \alpha (1 - 8n) - \frac{n}{1 + n^2}
$$

We are interested in finding the values of $\alpha$ for which the equation $F(n) = 0$ has either one or three real roots. This transition typically occurs at a critical point where the function $F(n)$ is tangent to the x-axis, meaning both $F(n) = 0$ and its derivative $F'(n) = 0$ at the same point.

First, let's find the derivative of $F(n)$ with respect to $n$:

$$
F(n) = \alpha (1 - 8n) - \frac{n}{1 + n^2}
$$

The derivative $F'(n)$ is:

$$
F'(n) = \alpha (-8) - \frac{(1 + n^2) - n(2n)}{(1 + n^2)^2}
$$

$$
F'(n) = -8\alpha - \frac{1 + n^2 - 2n^2}{(1 + n^2)^2}
$$

$$
F'(n) = -8\alpha - \frac{1 - n^2}{(1 + n^2)^2}
$$

For the critical point, we need both $F(n) = 0$ and $F'(n) = 0$:

1. $F(n) = \alpha (1 - 8n) - \frac{n}{1 + n^2} = 0$
2. $F'(n) = -8\alpha - \frac{1 - n^2}{(1 + n^2)^2} = 0$

Let's solve these equations simultaneously. From $F'(n) = 0$:

$$
-8\alpha - \frac{1 - n^2}{(1 + n^2)^2} = 0
$$

$$
8\alpha = -\frac{1 - n^2}{(1 + n^2)^2}
$$

Since $\alpha$ is positive, the right-hand side must also be positive, which implies:

$$
1 - n^2 < 0
$$

$$
n^2 > 1
$$

$$
  n| > 1
$$

Thus, we are interested in $n > 1$ or $n < -1$. However, since $n$ represents a population size, we consider $n > 1$.

Now, let's denote $n$ as the point where both $F(n) = 0$ and $F'(n) = 0$. We can use numerical methods to find the value of $n$ that satisfies both conditions.


```python
def F(n, alpha):
    return alpha * (1 - 8 * n) - n / (1 + n**2)

def F_prime(n, alpha):
    return -8 * alpha - (1 - n**2) / (1 + n**2)**2

# Find the critical alpha
def equations(p):
    n, alpha = p
    return [F(n, alpha), F_prime(n, alpha)]

initial_guess = [1.5, 0.5]
n_solution, alpha_solution = fsolve(equations, initial_guess)

print(f"Critical n: {n_solution:.4f}")
print(f"Critical alpha: {alpha_solution:.4f}")

# Generate n values
n_values = np.linspace(0, 2, 400)

# Plot F(n) for different alpha values around the critical alpha
alpha_values = np.linspace(0.2, 0.8, 10)
plt.figure(figsize=(10, 6))

for alpha in alpha_values:
    F_values = F(n_values, alpha)
    plt.plot(n_values, F_values, label=f'$\\alpha = {alpha:.2f}$')

# Plot the critical alpha
F_critical = F(n_values, alpha_solution)
plt.plot(n_values, F_critical, label=f'Critical $\\alpha = {alpha_solution:.4f}$', linestyle='--', linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(n_solution, color='red', linestyle=':', label=f'Critical $n = {n_solution:.4f}$')
plt.xlabel('$n$')
plt.ylabel('$F(n)$')
plt.title('Plot of $F(n)$ for different values of $\\alpha$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def F(n, alpha):
    return alpha * n * (1 - 8 * n) - n**2 / (1 + n**2)

def F_prime(n, alpha):
    return -8 * alpha - (1 - n**2) / (1 + n**2)**2

def find_fixed_points(alpha):
    # Initial guesses for the roots
    initial_guesses = [0.1, 0.5, 1.0, 1.5]
    fixed_points = []
    for guess in initial_guesses:
        root = fsolve(F, guess, args=(alpha,))
        # Avoid duplicate roots
        if not any(np.isclose(root, fp, atol=1e-3) for fp in fixed_points):
            fixed_points.append(root[0])
    return fixed_points

def determine_stability(fixed_points, alpha):
    stability = []
    for fp in fixed_points:
        derivative = F_prime(fp, alpha)
        if derivative < 0:
            stability.append("Stable")
        elif derivative > 0:
            stability.append("Unstable")
        else:
            stability.append("Non-hyperbolic")
    return stability

alpha_values = np.linspace(0.2, 0.8, 25)
results = []

for alpha in alpha_values:
    fixed_points = find_fixed_points(alpha)
    stability = determine_stability(fixed_points, alpha)
    results.append((alpha, fixed_points, stability))

for result in results:
    alpha, fixed_points, stability = result
    print(f"Alpha: {alpha:.4f}")
    for fp, stab in zip(fixed_points, stability):
        print(f"  Fixed Point: {fp:.4f}, Stability: {stab}")
    print()
```
