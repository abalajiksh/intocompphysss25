## 1. Fixed Points of the Lorenz System

```python
import numpy as np

# Define the coefficients of the characteristic polynomial
sigma = 10  # Example value for sigma
b = 8/3    # Example value for b
r = 28     # Example value for r

coeffs = [
    1,                # Coefficient for λ³
    1 + b + sigma,    # Coefficient for λ²
    b * (sigma + r),  # Coefficient for λ
    2 * sigma * b * (r - 1)  # Constant term
]

# Find the roots of the characteristic polynomial
eigenvalues = np.roots(coeffs)

print("Eigenvalues:", eigenvalues)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the characteristic polynomial P(λ)
def P(λ, σ, b, r):
    return λ**3 + (1 + b + σ) * λ**2 + b * (σ + r) * λ + 2 * σ * b * (r - 1)

# Parameters
σ = 10
b = 8/3
r_values = np.linspace(0, 1.8, 6)  # Discrete values of r from 0 to 1.8

# Range of λ values
λ_values = np.linspace(-20, 20, 400)

# Plot P(λ) for each r value
plt.figure(figsize=(12, 8))
for r in r_values:
    P_values = P(λ_values, σ, b, r)
    plt.plot(λ_values, P_values, label=f'r = {r:.2f}')

plt.axhline(0, color='black', linewidth=0.5)  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5)  # Vertical line at x=0
plt.xlabel('λ')
plt.ylabel('P(λ)')
plt.title('Characteristic Polynomial P(λ) for Different Values of r')
plt.legend()
plt.grid(True)
plt.show()
```
